# Fantasy Baseball Rankings

Uses the `Fantasy` class from `baseball.py` to rank all qualified MLB players by expected fantasy contribution under Yahoo standard public league scoring.

**Three scoring modes:**
- `points` — per-stat point values (e.g. HR = 10.4 pts)
- `categories` — 5×5 head-to-head categories, valued by z-score
- `roto` — 5×5 rotisserie, same z-score valuation

**Also covers:** FanGraphs projection systems (Steamer, ZiPS, Depth Charts, etc.) via the `.projections` property on `Batters`, `Pitchers`, and `League`.

## Section 4: Categories / Roto Rankings (Z-Score)

No literal point totals exist in category leagues — player value is expressed as
the sum of z-scores across all five batting or pitching categories.

Rate stats are weighted by playing time before z-scoring:
- `AVG` → `excess_hits = (AVG − pool_mean) × AB`
- `ERA` → `era_contrib = (pool_mean − ERA) × IP / 9`
- `WHIP` → `whip_contrib = (pool_mean − WHIP) × IP`

```python
f_cat = Fantasy(SEASON, 'categories', league=league)
df_cat = f_cat.rank()

print(f'Ranked {len(df_cat)} players')
print(f'Z-score columns: {[c for c in df_cat.columns if c.startswith("z_")]}')
```

## Yahoo Average Draft Position (ADP)

Fetches ADP from Yahoo's public draft analysis API (`out=draft_analysis`).
Sorted by `average_pick` ascending — lower = earlier pick.

**Requires:** `browser/yahoo_oauth.json` (run Section 2 of `yahoo_fantasy_baseball.ipynb` first).

```python
import logging
from yahoo_oauth import OAuth2

logging.getLogger('yahoo_oauth').setLevel(logging.WARNING)

CREDS_FILE = 'browser/yahoo_oauth.json'
_BASE      = 'https://fantasysports.yahooapis.com/fantasy/v2'
ADP_COUNT  = 300   # number of players to fetch

def _yflatten(data):
    """Collapse Yahoo list-of-single-key-dicts into one dict."""
    if isinstance(data, dict):
        return data
    out = {}
    for item in data:
        if isinstance(item, dict):
            out.update(item)
    return out

_oa = OAuth2(None, None, from_file=CREDS_FILE)
if not _oa.token_is_valid():
    _oa.refresh_access_token()

def _yget(path):
    if not _oa.token_is_valid():
        _oa.refresh_access_token()
    resp = _oa.session.get(f'{_BASE}{path}?format=json')
    resp.raise_for_status()
    return resp.json()

# Resolve game key for the target season (/games plural, game[0] is a plain dict)
_game_data = _yget(f'/games;game_codes=mlb;seasons={SEASON}')
_game_key  = _game_data['fantasy_content']['games']['0']['game'][0]['game_key']
print(f'Game key for {SEASON}: {_game_key}')

rows = []
for _start in range(0, ADP_COUNT, 25):
    _data    = _yget(f'/game/{_game_key}/players;sort=DA_AR;start={_start};count=25;out=draft_analysis')
    _players = _data['fantasy_content']['game'][1]['players']
    if not isinstance(_players, dict) or _players.get('count', 0) == 0:
        break
    for i in range(_players['count']):
        p        = _players[str(i)]['player']
        info     = _yflatten(p[0])
        da       = _yflatten((p[1] if len(p) > 1 else {}).get('draft_analysis', []))
        name_raw = info.get('name', '')
        rows.append({
            'name':        name_raw.get('full', '') if isinstance(name_raw, dict) else str(name_raw),
            'team':        info.get('editorial_team_abbr', ''),
            'positions':   info.get('display_position', ''),
            'adp':         float(da.get('average_pick',    999)),
            'avg_round':   float(da.get('average_round',   999)),
            'pct_drafted': round(float(da.get('percent_drafted', 0)) * 100, 1),
        })

yahoo_adp = (
    pd.DataFrame(rows)
    .sort_values('adp')
    .reset_index(drop=True)
)
yahoo_adp.index += 1
yahoo_adp.index.name = 'rank'

print(f'Yahoo ADP ({SEASON}) — {len(yahoo_adp)} players')
yahoo_adp
```

## Draft Value — Projected Points vs ADP

Players where projected points rank is significantly better than ADP rank are undervalued picks.
`value = adp_rank − pts_rank` — higher is better (drafted later than they should be).

```python
SCORING_TYPE = 'roto'   # 'points', 'categories', or 'roto'

import unicodedata
import re 

def _norm(name):
    """Normalize name for cross-source joins: lowercase, strip accents, strip Yahoo type suffix."""
    s = str(name).lower()
    s = re.sub(r'\s*\((batter|pitcher)\)', '', s)
    s = unicodedata.normalize('NFKD', s)
    return ''.join(c for c in s if not unicodedata.combining(c)).strip()

# ── Select rankings DataFrame based on scoring type ──────────────────────────
_df_map = {'points': df_pts, 'categories': df_cat, 'roto': df_roto}
if SCORING_TYPE not in _df_map:
    raise ValueError(f"SCORING_TYPE must be one of {list(_df_map)}")
_rank_df = _df_map[SCORING_TYPE]

# ── ADP table (top 300 from Yahoo) ───────────────────────────────────────────
adp = (
    yahoo_adp.reset_index()
    .rename(columns={'rank': 'adp_rank'})
    [['adp_rank', 'name', 'positions', 'adp', 'avg_round']]
)
adp['_key'] = adp['name'].apply(_norm)

# ── Fantasy rankings table ────────────────────────────────────────────────────
_score_label = {'points': 'proj_pts', 'categories': 'cat_zscore', 'roto': 'roto_zscore'}[SCORING_TYPE]
rnk = (
    _rank_df.reset_index()
    .rename(columns={'rank': 'fantasy_rank', 'fantasy_score': _score_label})
    [['fantasy_rank', 'Name', _score_label, 'player_type']]
)
rnk['_key'] = rnk['Name'].apply(_norm)

# ── Join on normalized name ───────────────────────────────────────────────────
merged = adp.merge(rnk, on='_key')

# value: positive = drafted later than rankings suggest → undervalued
merged['value'] = merged['adp_rank'] - merged['fantasy_rank']

value_df = (
    merged[['name', 'player_type', 'positions', 'adp', 'adp_rank', 'fantasy_rank', _score_label, 'value']]
    .sort_values('value', ascending=False)
    .reset_index(drop=True)
)
value_df.index += 1
value_df.index.name = 'rank'

print(f'Scoring: {SCORING_TYPE}  |  {len(value_df)} matched  |  {len(adp) - len(value_df)} unmatched')
print(f'\nTop 30 undervalued — drafted much later than {SCORING_TYPE} rankings suggest:')
value_df.head(30)
```
