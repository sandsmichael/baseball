# Fantasy Baseball Rankings

Uses the `Fantasy` class from `baseball.py` to rank all qualified MLB players by expected fantasy contribution under Yahoo standard public league scoring.

**Three scoring modes:**
- `points` — per-stat point values (e.g. HR = 10.4 pts)
- `categories` — 5×5 head-to-head categories, valued by z-score
- `roto` — 5×5 rotisserie, same z-score valuation

**Also covers:** FanGraphs projection systems (Steamer, ZiPS, Depth Charts, etc.) via the `.projections` property on `Batters`, `Pitchers`, and `League`.

### Categories / Roto Rankings (Z-Score)

No literal point totals exist in category leagues — player value is expressed as
the sum of z-scores across all five batting or pitching categories.

Rate stats are weighted by playing time before z-scoring:
- `AVG` → `excess_hits = (AVG − pool_mean) × AB`
- `ERA` → `era_contrib = (pool_mean − ERA) × IP / 9`
- `WHIP` → `whip_contrib = (pool_mean − WHIP) × IP`

```python
f_cat = Fantasy(SEASON, 'categories', league=league)
df_cat = f_cat.rank()

print(f'Ranked {len(df_cat)} players')
print(f'Z-score columns: {[c for c in df_cat.columns if c.startswith("z_")]}')
```

## FanGraphs Projections

`.projections` is a lazy property on `Batters`, `Pitchers`, and `League` — fetched from the FanGraphs public JSON API on first access, then cached.

Available systems: `steamer`, `steamerr`, `zips`, `zipr`, `atc`, `thebat`, `thebatx`, `fangraphs` (Depth Charts).

> **Note:** `fangraphs` (Depth Charts) and `zipr` are preseason-only — they return errors for completed seasons. Use `steamer`, `zips`, `atc`, `thebat`, or `thebatx` for historical data.

### Fantasy Score Estimate based on Projections

```python
SCORING_TYPE = 'categories'  # 'points', 'categories', or 'roto'
PROJ_SYSTEM  = 'atc'

league = League(SEASON, proj_system=PROJ_SYSTEM)
f = Fantasy(SEASON, SCORING_TYPE, league=league, use_projections=True)
df = f.rank()

score_cols = [c for c in df.columns if c.startswith(('pts_', 'z_'))]
df[['Name', 'Pos', 'fantasy_score'] + score_cols].sort_values('fantasy_score', ascending=False).iloc[:55]
```

## Yahoo Average Draft Position (ADP)

Fetches ADP from Yahoo's public draft analysis API (`out=draft_analysis`).
Sorted by `average_pick` ascending — lower = earlier pick.

**Requires:** `browser/yahoo_oauth.json` (run Section 2 of `yahoo_fantasy_baseball.ipynb` first).

```python
import logging
import sys; sys.path.insert(0, '..')
from yahoo import Yahoo

logging.getLogger('yahoo_oauth').setLevel(logging.WARNING)

CREDS_FILE = '../browser/yahoo_oauth.json'
ADP_COUNT  = 300
ADP_SEASON = 2026  # upcoming season — ADP is pre-season draft data, separate from SEASON

yf = Yahoo(season=ADP_SEASON, creds_file=CREDS_FILE)
yahoo_adp = yf.adp(ADP_COUNT)

print(f'Yahoo ADP ({ADP_SEASON}) — {len(yahoo_adp)} players')
yahoo_adp
```

## Draft Value — Projected Points vs ADP

Players where projected points rank is significantly better than ADP rank are undervalued picks.
`value = adp_rank − pts_rank` — higher is better (drafted later than they should be).

```python
SCORING_TYPE = 'roto'   # 'points', 'categories', or 'roto'

_df_map = {'points': df_pts, 'categories': df_cat, 'roto': df_roto}
if SCORING_TYPE not in _df_map:
    raise ValueError(f"SCORING_TYPE must be one of {list(_df_map)}")

value_df = yf.draft_value(_df_map[SCORING_TYPE])

print(f'Scoring: {SCORING_TYPE}  |  {len(value_df)} matched  |  {ADP_COUNT - len(value_df)} unmatched')
print(f'\nTop 30 undervalued — drafted much later than {SCORING_TYPE} rankings suggest:')
value_df.head(30)
```

## Projection Targets: Power/Speed & Average

```python
bat_proj = league.batters.projections

SHOW = ['PlayerName', 'Team', 'G', 'PA', 'HR', 'R', 'RBI', 'SB', 'AVG']
avail = [c for c in SHOW if c in bat_proj.columns]

print('── 20/20 Club (20+ HR, 20+ SB) ─────────────────────────────────────')
club_20_20 = (bat_proj[(bat_proj.HR >= 20) & (bat_proj.SB >= 20)]
              .sort_values(['HR', 'SB'], ascending=False)[avail])
display(club_20_20.round({'HR': 1, 'R': 1, 'RBI': 1, 'SB': 1, 'AVG': 3}))

print(f'\n── 30/30 Club (30+ HR, 30+ SB) ─────────────────────────────────────')
club_30_30 = (bat_proj[(bat_proj.HR >= 30) & (bat_proj.SB >= 30)]
              .sort_values(['HR', 'SB'], ascending=False)[avail])
display(club_30_30.round({'HR': 1, 'R': 1, 'RBI': 1, 'SB': 1, 'AVG': 3}))

print(f'\n── .295+ AVG ────────────────────────────────────────────────────────')
avg_295 = (bat_proj[bat_proj.AVG >= 0.295]
           .sort_values('AVG', ascending=False)[avail])
display(avg_295.round({'HR': 1, 'R': 1, 'RBI': 1, 'SB': 1, 'AVG': 3}))
```
